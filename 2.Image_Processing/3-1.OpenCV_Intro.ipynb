# CoLab에서 cv2.imshow() 주의할 점


```
print(cv2.IMREAD_COLOR,cv2.IMREAD_GRAYSCALE,cv2.IMREAD_UNCHANGED)
img_cv = cv2.imread('f.png',cv2.IMREAD_UNCHANGED)

## CoLab에서만 작동안함.
# cv2.imshow('xx',img_cv)

from google.colab.patches import cv2_imshow
# cv2_imshow(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB))
cv2_imshow(img_cv )
```

# CoLab에서 plt.imshow는?
* 잘 작동함.


```
from matplotlib import pyplot as plt

img_plt = plt.imread('f.png')
plt.imshow(img_plt)
```

# 크기 변환


```

cv2_imshow(cv2.resize(img_cv, (256, 256)) )
plt.imshow(cv2.resize(img_plt,(256,256)))
```
