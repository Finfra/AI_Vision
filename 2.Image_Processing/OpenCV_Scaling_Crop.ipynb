# Resize

```
import cv2
from google.colab.patches import cv2_imshow
filename = "elephant.png"
![ ! -f {filename} ]&&wget https://raw.githubusercontent.com/Finfra/AI_Vision/master/data/img/{filename}
img = cv2.imread(filename)

print(img.shape)
w,h = int(img.shape[1]/4),int(img.shape[0]/4) 

img_new = cv2.resize(img,(w,h))

print(img_new.shape)

cv2_imshow(img)
cv2_imshow(img_new)

```
