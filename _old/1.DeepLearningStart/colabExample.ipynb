```
import numpy as np
import matplotlib.pyplot as plt

# 검정색으로 채우기
i=np.zeros((10,10,4))
# 4번째 값은 알파값(1로 채워서 값을 적용하기)
i[:,:,3]=1.
# 세로로 빨간색 집어 넣기
i[:,:,0]=np.array([0,1]*50).reshape((10,10))
# 가로로 녹색 집어 넣기
i[range(1,10,2),:,1]=1

plt.imshow(i, cmap='Greys')
plt.show()

plt.imsave('a.png',i)

ir=plt.imread('a.png')
print(ir)
```
