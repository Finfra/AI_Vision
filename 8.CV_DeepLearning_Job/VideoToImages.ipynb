```
import cv2
![ ! -f chaplin.mp4 ]&&wget https://raw.githubusercontent.com/spmallick/learnopencv/master/tracking/videos/chaplin.mp4

cam = cv2.VideoCapture('chaplin.mp4')
currentFrame = 0
while(True):
    ret, frame = cam.read()
    if ret:
        cv2.imwrite(str(currentFrame) + '.jpg', frame)
        currentFrame += 1
    else:
        break
cam.release()
```
